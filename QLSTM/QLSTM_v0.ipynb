```python
import torch
import torch.nn as nn
import pennylane as qml
import numpy as np

from QLSTM_v0 import VQC, CustomQLSTMCell, CustomLSTM


MODEL_PATH = "QLSTM_TS_MODEL_DAMPED_SHM_1/QLSTM_TS_MODEL_DAMPED_SHM_1_NO_1_Epoch_100_torch_model.pth"
input_size = 1
hidden_size = 5
output_size = 1
qnn_depth = 5

qlstm_cell = CustomQLSTMCell(input_size, hidden_size, output_size, qnn_depth).float()
model = CustomLSTM(input_size, hidden_size, qlstm_cell).float()

print("[success] model created")
print(model)

model.load_state_dict(torch.load(MODEL_PATH))
print(f"[success] trained weight loaded")
model.eval()

# create fake for predicting
sample_input = torch.randn(1, 4, input_size).float() # 1筆資料, 序列長度4, 特徵維度1

with torch.no_grad():
    prediction, _ = model(sample_input)

# get prediction from last time step
last_time_step_prediction = prediction[:, -1, :]

print(f"input shape: {sample_input.shape}")
print(f"outout shape: {prediction.shape}")
print(f"prediction: {last_time_step_prediction.item()}")
```
